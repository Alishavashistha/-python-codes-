```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples and features
num_samples = 100
num_features = 3

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, num_features)

# Define true weights (slopes) and intercept
true_weights = np.array([[3], [2], [1]])  # shape: (3, 1)
true_intercept = 4

# Generate noise
noise = np.random.randn(num_samples, 1)

# Compute y values
y_train = true_intercept + x_train @ true_weights + noise  # @ is matrix multiplication

# Plotting using only first feature for visualization
plt.figure(figsize=(8, 5))
plt.scatter(x_train[:, 0], y_train, color="green", label="Training data (vs 1st feature)", alpha=0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:", y_train.shape)
```

```python
#Initialize  number of samples, weight and bias
m = x_train.shape[0]
w = np.array([[0.5],[0.3],[0.2]])
b=0

#number of iterations and learning rate for gradient descent
iteration = 1000
alpha = 0.01
cost = []    #store cost values

# Gradient Descent Loop
for i in range(1000):
    y_pred = x_train@w+b
    j = (1/2*m) * np.sum((y_pred - y_train) ** 2)
    cost.append(j)
    # Compute gradients and update weights and bias
    w= w - alpha* (1/m) * (x_train.T @ (y_pred - y_train))
    b = b - alpha * (1/m) * np.sum(y_pred - y_train)

# Print final learned weights and bias
print(w)
print(b)

# Plot cost over iterations to visualize convergence
plt.plot(cost)
plt.show()

# Print final cost value
print(cost[-1])
```
